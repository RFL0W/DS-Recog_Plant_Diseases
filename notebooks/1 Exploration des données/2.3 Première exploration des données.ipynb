### Elements graphiques pour soutenance

```python
# Change the default font to Calibri
plt.rcParams['font.family'] = 'Calibri'

# Define the colors in RGB format
bar_color1 = (122/255, 176/255, 110/255)
bar_color2 = (120/255, 158/255, 192/255)

# Filter the data
df_filtered = df_KaggleData[(df_KaggleData['Set'] != 'grayscale') & (df_KaggleData['Set'] != 'segmented') & (df_KaggleData['Dataset'] != 'PlantDisease')]

# Create a countplot
plt.figure(figsize=(10, 3))  # adjust the size of the plot as needed
ax = sns.countplot(data = df_filtered, x = 'Categorie', hue = 'Dataset', palette=[bar_color1, bar_color2])

# Rotate the category names by 45 degrees
plt.xticks(rotation=45, horizontalalignment='right')

# Remove the box around the plot
sns.despine()

# Set the labels for the axes and make them bold
ax.set_xlabel('Categorie', weight='bold', fontsize = 18)
ax.set_ylabel('Nombre d\'images', weight='bold', fontsize = 18)

# Remove the legend
ax.get_legend().remove()

# Show the plot
plt.show()
```

```python
# Group the filtered data by 'Categorie' and 'Dataset' and count the number of values in each group
grouped_counts = df_filtered.groupby(['Categorie', 'Dataset']).size().reset_index(name='Count')

# Get the unique dataset names
datasets = grouped_counts['Dataset'].unique()

# For each dataset, print the min and max count
for dataset in datasets:
    min_count = grouped_counts[grouped_counts['Dataset'] == dataset]['Count'].min()
    max_count = grouped_counts[grouped_counts['Dataset'] == dataset]['Count'].max()
    print(f"For dataset {dataset}, min count is {min_count} and max count is {max_count}")
```
