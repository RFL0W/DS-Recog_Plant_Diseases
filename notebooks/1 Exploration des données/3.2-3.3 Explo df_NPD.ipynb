**Graphiques sur les potentielles augmentations de données**

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(data = df_NPD[df_NPD['Suffixe']!=''], x = 'Suffixe')
ax.set_title("Nombre d'images ajoutées par type de transformations")
ax.set_xlabel("Type de transformation")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right');
```

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(data = df_NPD[df_NPD['Rotation']!=0], x = 'Rotation')
ax.set_title("Nombre d'images avec rotation")
ax.set_xlabel("Degré de rotation");
```

```python
ax = sns.countplot(data = df_NPD[df_NPD['Symetrie']!=0], x = 'Symetrie')
ax.set_title("Nombre d'images avec symétrie")
ax.set_xticks(ticks=[0,1], labels=['Gauche-Droite', 'Haut-Bas'])
ax.set_xlabel("Type de Symétrie");
```

## Répartition des différentes catégories

```python
plt.figure(figsize=(6, 6))

plt.pie(x=df_NPD['Saine'].value_counts(),
        labels=['Malades', 'Saines'],
        colors=['#b42253','#d2a3b3'],
        explode=[0, 0.1],
        autopct='%1.1f%%',
        pctdistance=0.7, 
        labeldistance=1.1,
        shadow=True)

plt.title("Part de feuilles saines ou malades")
plt.legend();
```

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(data = df_NPD, x = 'Plante',hue='Saine')
ax.set_title("Nombre d'images saines ou malades par type de plante")
ax.set_xlabel("Type de plante")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right');
```

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(data = df_NPD[df_NPD['Rotation']!=0], x = 'Categorie', hue='Rotation')
ax.set_title("Nombre d'images avec rotation par catégorie de feuilles")
ax.set_xlabel("Catégorie de feuilles")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right');
```

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(data = df_NPD, x = 'Categorie', hue='Rotation')
ax.set_title("Nombre d’images avec rotation par catégorie de feuilles (avec les rotations nulles)")
ax.set_xlabel("Catégorie de feuilles")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right');
```

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(data = df_NPD[df_NPD['Symetrie']!=0], x = 'Categorie', hue='Symetrie')
ax.set_title("Nombre d'images avec symétrie par catégorie de feuilles")
ax.set_xlabel("Catégorie de feuilles")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right');
```

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(data = df_NPD[((df_NPD['Symetrie']!=0) | (df_NPD['Rotation']!=0))], x = 'Categorie')
ax.set_title("Nombre d'images avec symétrie ou rotation par catégorie de feuilles")
ax.set_xlabel("Catégorie de feuilles")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right');
```

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(data = df_NPD[((df_NPD['Symetrie']==0) & (df_NPD['Rotation']==0))], x = 'Categorie')
ax.set_title("Nombre d'images sans symétrie et rotation par catégorie de feuilles")
ax.set_xlabel("Catégorie de feuilles")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right');
```

## Eléments graphiques pour soutenance

```python
# Change the default font to Calibri
plt.rcParams['font.family'] = 'Calibri'

# Define the color in RGB format
bar_color = (50/255, 80/255, 106/255)

# Count the occurrences of each category and sort by category
category_counts = df_NPD['Categorie'].value_counts().sort_index()

# Create a dataframe from the series for the countplot
category_counts_df = pd.DataFrame({'Categorie': category_counts.index, 'Count': category_counts.values})

# Create a countplot
plt.figure(figsize=(10, 3))  # adjust the size of the plot as needed
ax = sns.barplot(x='Categorie', y='Count', data=category_counts_df, color=bar_color)

# Rotate the category names by 45 degrees
plt.xticks(rotation=45, horizontalalignment='right')

# Remove the box around the plot
sns.despine()

# Set the labels for the axes and make them bold
ax.set_xlabel('Categorie', weight='bold', fontsize = 18)
ax.set_ylabel('Nombre d\'images', weight='bold', fontsize = 18)


# Show the plot
plt.show()
```

```python
# Change the default font to Calibri
plt.rcParams['font.family'] = 'Calibri'

# Define the color in RGB format
bar_color = (120/255, 158/255, 192/255)

# Count the occurrences of each category and sort by category
category_counts = df_NPD['Saine'].value_counts().sort_index()

# Create a dataframe from the series for the countplot
category_counts_df = pd.DataFrame({'Saine': category_counts.index, 'Count': category_counts.values})

# Replace 0 and 1 in 'Saine' with 'Malade' and 'Saine' respectively
category_counts_df['Saine'].replace({0: 'Malade', 1: 'Saine'}, inplace=True)

# Create a countplot
plt.figure(figsize=(5, 3))  # adjust the size of the plot as needed
ax = sns.barplot(x='Saine', y='Count', data=category_counts_df, color=bar_color)

# Remove the box around the plot
sns.despine()

# Set the labels for the axes and make them bold
ax.set_xlabel('Santée', weight='bold', fontsize = 18)
ax.set_ylabel('Nombre d\'images', weight='bold', fontsize = 18)

# Show the plot
plt.show()
```
